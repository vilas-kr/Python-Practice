```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Ram", "Sam", "John", "Ram", "Sam"],
    "Dept": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [1000, 2000, 3000, np.nan, 2000],
    "Age": [22, 25, 28, 22, np.nan]
})
```

```python
# Find:
# First 3 rows
# Last 2 rows
# Shape of DataFrame
# Column names

print(df.head(3))
print(df.tail(2))
print(df.shape)
print(df.columns)
```

```python
# Task 2: Data Cleaning

# Count missing values
# Fill missing Salary with mean
# Fill missing Age with median

print(df.isnull().sum())
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df["Age"] = df["Age"].fillna(df["Age"].median())
print(df)
```

```python
# Task 3: Filtering
# Employees with Salary > 1500
# Employees from IT department

high_sal = df[df["Salary"] > 1500]
it_emp = df[df["Dept"] == "IT"]
print(high_sal)
print(it_emp)
```

```python
# Task 4: Column Operations
# Create:
# New column:
# Bonus = Salary * 0.1

df["Bonus"] = df["Salary"] * 0.1
print(df)
```

```python
# Task 5: GroupBy
# Find:
# Total salary per Dept
# Average salary per Dept

dept_sal = df.groupby("Dept")["Salary"].agg(["sum", "mean"])
print(dept_sal)
```

```python
# Task 6: Duplicates
# Remove duplicate rows

df = df.drop_duplicates()
print(df)
```

```python
# Task 7: Sorting
# Sort:
# By Salary descending
print(df.sort_values("Salary", ascending=False))
```

```python
# Create column:
# "Level"
# Salary > 2000 → High
# Else → Low

df["level"] = np.where(df["Salary"] > 2000, 'High', 'Low')
print(df)
```

```python
# Find:
# Dept with highest average salary
dept_avg_sal = df.groupby("Dept")["Salary"].sum().reset_index()
print(dept_avg_sal.iloc[dept_avg_sal["Salary"].argmax(), 0])
```

```python
# Drop rows where Salary is missing
# Then group by Dept and get total salary

df = df.dropna(subset=["Salary"])
dept_total_sal = df.groupby("Dept")["Salary"].sum()
print(dept_total_sal)
```
