```python
import pandas as pd
import numpy as np

emp = pd.DataFrame({
    "EmpID": [1, 2, 3, 4, 5],
    "Name": ["Ram", "Sam", "John", "Ravi", "Anu"],
    "Dept": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [1000, 2000, 3000, 4000, np.nan]
})
```

```python
dept = pd.DataFrame({
    "Dept": ["IT", "HR"],
    "Manager": ["Alice", "Bob"]
})
```

```python
# Merge both tables to get:
# EmpID, Name, Dept, Salary, Manager
emp_dept = pd.merge(emp, dept, on="Dept", how="inner")
print(emp_dept)
```

```python
# Fill missing Salary with Dept average salary
emp_dept["Salary"] = emp_dept["Salary"].fillna(
    emp_dept.groupby("Dept")["Salary"].transform("mean")
)
print(emp_dept)
```

```python
# Employees whose Salary > dept average salary
emp_dept["Dept_Avg"] = emp_dept.groupby("Dept")["Salary"].transform("mean")
result = emp_dept[emp_dept["Salary"] > emp_dept["Dept_Avg"]]
print(result)
```

```python
# Total salary per Manager
mng_total_sal = emp_dept.groupby("Manager")["Salary"].sum()
print(mng_total_sal)
```

```python
# Rank employees based on Salary (highest = 1)
salary_rank = emp_dept.sort_values("Salary", ascending=False).reset_index()
salary_rank["Rank"] = salary_rank.index
print(salary_rank)
```

```python
# Create column:
# "Category"
# Salary > 3000 → High
# 2000–3000 → Medium
# < 2000 → Low

emp_dept["Category"] = np.where(emp_dept["Salary"] > 3000, 'high', 
        np.where(emp_dept["Salary"] > 2000, 'medium', 'low')
    )
print(emp_dept)
```

```python
# Dept where total salary > 5000
dept_avg = emp_dept.groupby("Dept")["Salary"].sum().reset_index()
print(dept_avg[dept_avg["Salary"] > 5000])
```
