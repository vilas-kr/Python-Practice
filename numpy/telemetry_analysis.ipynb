```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("telemetry_data.csv")
```

```python
# Extract temperature as NumPy array
temps = df['temperature'].values
```

```python
# Compute: mean, std, min, max 
max_val = np.max(temps)
min_val = np.min(temps)
std_val = np.std(temps)
mean_val = np.mean(temps)
print(f"mean : {mean_val}, max: {max_val}, min: {min_val}, std: {std_val}")
```

```python
# Normalize to: normalized = (temps - min_val) / (max_val - min_val)
temps_normalize = (temps - min_val) / (max_val - min_val)
```

```python
# Mask anomalies: mask = temps > mean + 2 * std
mask = np.argwhere(temps > (mean_val + 2 * std_val))
```

```python
total_count = np.sum(mask)
```

```python
temps_normalize[mask] = np.nan
```

```python
#  compute mean temperature and battery per device
device_stats = df.groupby("device_id")[['temperature', 'battery_level']].mean()
device_stats
```

```python
df['temp_rolling_mean'] = df['temperature'].rolling(window=5).mean()
df.head(10)
```

```python
# Filter high-risk: devices with mean temp >25°C AND any battery <0.3
high_risk = df[(df.groupby('device_id')['temperature'].mean() > 25) & (df['battery_level'] < 0.3)]
high_risk
```
