[[1, 2, np.nan],

 [7, 8, 9]]

Perform:
Replace NaN with column-wise mean
Find row with maximum sum
Normalize array (0 to 1 scale):
(𝑥−𝑚𝑖𝑛)/(𝑚𝑎𝑥−𝑚𝑖𝑛)
Find indices where values > 0.5 after normalization
Flatten and remove duplicates

```python
import numpy as np
```

```python
arr = np.array([[1, 2, np.nan],
                [4, np.nan, 6],
                [7, 8, 9]])
```

```python
# Replace NaN with column-wise mean
col_mean = np.nanmean(arr, axis=0)
print(col_mean)
inds = np.where(np.isnan(arr))
print(inds)
arr[inds] = np.take(col_mean, inds[1])
print(arr)
```

```python
# Find row with maximum sum
max_sum = np.sum(arr, axis=1)
print(np.argmax(max_sum))
```

```python
# Normalize array (0 to 1 scale): (𝑥−𝑚𝑖𝑛)/(𝑚𝑎𝑥−𝑚𝑖𝑛)
nrm_arr = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))
print(nrm_arr)
```

```python
# Find indices where values > 0.5 after normalization
max_ind = np.argwhere(nrm_arr > 0.5)
print(max_ind)
```

```python
# Flatten and remove duplicates
flat = nrm_arr.flatten()
uq = np.unique(nrm_arr)
print(uq)
```
