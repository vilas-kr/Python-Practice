```python
import numpy as np
```

```python
arr = np.array([2, 4, 6, 8, 10])

# Multiply all elements by 3
mul_arr = arr * 3
print(mul_arr)
```

```python
# Replace elements greater than 15 with -1
mul_arr[mul_arr > 15] = -1
print(mul_arr)
```

```python
# Print indices of elements equal to -1
print(np.argwhere(mul_arr == -1))
```

```python
# Use np.where() to create a new array:
# if value > 5 → 1
# else → 0

new_arr = np.where(arr > 5, 1, 0)
print(new_arr)
```
