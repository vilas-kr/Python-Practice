```python
# Problem:
# [10, 20, np.nan, 40, np.nan, 60]

# Perform:
# Detect missing values
# Replace missing values with mean of array
# Remove missing values
# Find sum ignoring NaN
# Replace values greater than 30 with 999
```

```python
import numpy as np
```

```python
arr = np.array([10, 20, np.nan, 40, np.nan, 60])
```

```python
# Detect missing values
print(arr[np.isnan(arr)])
```

```python
# Replace missing values with mean of array
copy_arr = np.copy(arr)
copy_arr[np.isnan(copy_arr)] = np.nanmean(copy_arr)
print(copy_arr)
```

```python
# Remove missing values
clean_arr = arr[~np.isnan(arr)]
print(clean_arr)
```

```python
# Find sum ignoring NaN
print(np.nansum(arr))
```

```python
# Replace values greater than 30 with 999
arr = np.where(arr > 30, 999, arr)
print(arr)
```
