```python
# Create a NumPy array:
# [5, 10, 15, 20, 25]

# Then:
# * Print elements greater than 10
# * Replace elements greater than 20 with 100
# * Find the mean of the array
```

```python
import numpy as np
```

```python
arr = np.array([5, 10, 15, 20, 25])
```

```python
# Print elements greater than 10
print(arr[arr > 10])
```

```python
# Replace elements greater than 20 with 100
arr[arr > 20] = 100
```

```python
# Find the mean of the array
print(np.mean(arr))
```
